```python
import pandas as pd
from fim import apriori
```

```python
with open('articles.txt',encoding='latin-1') as f:
    tracts = [i.strip().split(" ") for i in f.readlines()]
```

```python
tracts
```

```python
ans1 = apriori(tracts, target='r', supp=0.03, conf=70, zmin=2)
ans1
```

```python
with open('rules.txt', 'w') as f:
    f.write(f'{ans1}')
```
