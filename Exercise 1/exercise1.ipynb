```python
import pandas as pd
import json
from collections import defaultdict
import operator
```

```python
# Reading data as a txt file
data = pd.read_csv("wiki_edit.txt",delimiter= " ", header = None)
data.columns = ['RevisionId', 'ArticleId', 'Timestamp', 'Editor']
data
```

```python
# Top 5 articles received highest number of edits.
data.groupby('ArticleId')['RevisionId'].count().nlargest(5)
```

```python
# Top 5 editors with most number of article edits.
data.iloc[:,[1,3]].drop_duplicates().groupby('Editor')['ArticleId'].count().nlargest(5) # dropping duplicate articleids 
```

```python
# 15 most frequent terms appearing on tweet messages. 
f = open('wh.json') # reading data from json file and then closing the connection
tweets = [json.loads(line) for line in f]
f.close()
tweets
```

```python
count = defaultdict(int) # assigning null integer value
for tweet in tweets:
    words = tweet['text'].split(' ')
    for word in words:
        count[word] += 1 # count for words/terms
 
```

```python
max_word = sorted(count.items(), key = operator.itemgetter(1))
max_word.reverse()
print("Top 15 most frequent appearing terms on twitter messages")
max_word[0:15] 
```
