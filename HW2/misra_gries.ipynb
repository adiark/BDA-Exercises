```python
data = pd.read_csv("wiki_edit2.txt",delimiter= " ", header = None)
data.columns = ['RevisionId', 'ArticleId', 'Timestamp', 'Editor']
```

```python
def misras_gries(array,k=1000):
    """Misra-Gries algorithm
    Keyword arguments:
    array -- list of integers
    k -- value of k (default 1000)
    """
    keys = {}
    for i in array:
        val = str(i)
        if val in keys:
            keys[val] = keys[val] + 1

        elif len(keys) < k - 1:
            keys[val] = 1

        else:
            for key in list(keys):
                keys[key] = keys[key] - 1
                if keys[key] == 0:
                    del keys[key]
    return keys
```

```python
misras_gries(data.ArticleId)
```
